* Customer Tenure 

```python
#

df["Customer_Since"] = (pd.Timestamp("today") - df["Dt_Customer"]).dt.days
```


*in above cell
*You are calculating how many days have passed since the customer joined.

→ pd.Timestamp("today")

        •This gets today’s date.

→ df["Dt_Customer"]

        •This is the date when the customer joined / became a customer.

→ pd.Timestamp("today") - df["Dt_Customer"]

        •This calculates the difference between today and the customer’s joining date.

        •The result is a time duration (Timedelta).

→.dt.days

        •Converts that time difference into number of days.

→df["Customer_Since"] = ...

        •Stores the number of days in a new column called Customer_Since.

# Clustering
* We are going to use k-means algorithm for clustering and we need to give a 'k' like the grouping number which is the number of  cluster.

* there is a way to determine the optimal numbers of clusters called elbow mehtod we are going to apply that

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10), wcss,marker = "o")
plt.title("Elbow Mehtos for Optimal K")
plt.xlabel("WCSS")
plt.show()
```

```python
# both 4 and 6 will work in this case but I am going to tak e 6 i think it it is better 

kmeans = KMeans(n_clusters = 6)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
df.head()
```

```python
cluster_summary = df.groupby("Cluster")[features].mean()
```

```python
cluster_summary
```

```python
df["Cluster"].value_counts()
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled)
df["PCA1"], df["PCA2"] = pca_data[:,0], pca_data[:,1] 
```

```python
pca_data
```

```python
sns.scatterplot(x="PCA1" , y="PCA2", hue = "Cluster", data = df, palette = "Set1")
plt.title("Customer Segmentation (PCA)")
plt.show()
```
